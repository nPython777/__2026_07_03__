## 容器型別提示(Collection Type Hints) 容器-串列([])(List)/元組([])(Tuple)/集合(Set)/字典({})(Dict)

```python
def get_full_name(first_name:str,last_name:str) -> str:
    full_name:str = first_name.title() + " " + last_name.title()
    return full_name
```

```python
def get_name_with_age(name:str,age:int) -> str:
    name_with_age= name + "is this old " + str(age)
    return name_with_age
```

```python
prices:dict[str, float] = {"蘋果": 35.0, "香蕉": 15.5}
```

```python
def process_items(items:list[str]) -> None:
    for item in items:
        print(item.upper())
```

```python
def print_prices(prices:dict[str, float]) -> None:
    for item_name, item_price in prices.items():
        print(f"商品:{item_name}, 價格:{item_price}")
```

```python
def process_data(items_t:tuple[int, int, str], items_s:set[bytes]) -> tuple[tuple[int, int, str],set[bytes]]:
    return items_t , items_s
```

## 複合與可選型別提示(Union & Optioanal)

```python
def process_item(item: int | str) -> None:
    print(f"處理資料:{item}")
```

```python
#舊的程式寫法
from typing import Union
def process_item1(item:Union[int, str]) -> None:
    print(item)
```

```python
def say_hi(name: str | None = None) -> None:
    if name is None:
        print("哈囉，世界")
    else:
        print(f"Hello! {name}")
say_hi()
say_hi("robert")
```

```python
#舊的程式寫法
from typing import Optional
def say_hi1(name: Optional[str] = None) -> None:
    if name is None:
        print("哈囉，世界")
    else:
        print(f"Hello! {name}")
say_hi1()
say_hi1("robert")
```

### 自訂class的語法-目的建立實體 這段程式示範了 Python 中 類別 (Class)、物件 (Object)、建構子 (__init__)、型別提示 (Type Hint) 以及 函式傳遞物件 的用法。

```python
class Person(object):
    def __init__(self, name:str):
        self.name = name

john: Person = Person(name="john")

def get_person_name(one_person: Person) -> str:
    return one_person.name

get_person_name(john)
```

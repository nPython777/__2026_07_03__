```python
from pydantic import BaseModel,Field,field_validator
from datetime import datetime

class AirSite(BaseModel):
    aqi:int
    county:str
    date:datetime = Field(alias="datacreationdate")
    lat:float = Field(alias = "latitude")
    lon:float = Field(alias="longitude")
    pm25:float = Field(alias="pm2.5")
    pollutant:str 
    site_name:str = Field(alias="sitename")
    status:str  

    @field_validator('aqi','lat','lon','pm25', mode='before')
    @classmethod
    def empty_to_zero(cls, v):
        return 0 if v == '' else v
```

```python
class Root(BaseModel):
    status:bool = True
    sites:list[AirSite]
```

```python
root = Root(sites=[AirSite(**item) for item in contents])
```
